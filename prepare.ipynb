# generate dictionary for old vs new names

```python
#4,01-001
#27-007-D2

import csv

import os
dir_current= os.getcwd()


dict_correction={}
samples=set()

file=open('sample.name.correction.csv')
reader=csv.reader(file)

next(reader, None)

for line in reader:
    old=line[0].split('_S')[0]
    #day_old=old.split('-D')[1]
    new=line[1]
    #day_new=new.split('-D')[1]
    
    #sample_new=new.split('-D')[0]
    #sample_old=old.split('-D')[0]
    
    samples.add(old)
    
    
    
    #dict_correction[(sample_new,day_new)]=(sample_old,day_old)
    dict_correction[old]=(new)
    
print dict_correction    
print len(samples),"samples were incorectly names bu UCLA core and will be renamed. The phenotype file is correct"
```

# rename directories

```python
import os 
from glob import glob



l_dir=glob("./salmon.counts/*")


for d in l_dir:
    old=d.split("/")[2].split('_S')[0]
    new=""
    if old in samples:
        new=dict_correction[old]    
    else:
        new=old
        
    
    
    cmd="mv "+dir_current+"/"+"salmon.counts/"+d.split("/")[2]+" "+ dir_current+"/salmon.counts/"+new
    os.system(cmd)
```

```python
day="day0"
import pandas as pd
data0=pd.read_csv('phenotype.mortality-twbga.day0.DE.csv',sep='\t')
phenotypes= set(data0["condition"].values)

print len(phenotypes)

data_twbga_lt_150_M0=data0.loc[data0['condition'] == 'twbga_lt_150_M0']
data_twbga_lt_150_M1=data0.loc[data0['condition'] == 'twbga_lt_150_M1']
data_twbga_gt_150_M1=data0.loc[data0['condition'] == 'twbga_gt_150_M1']
data_twbga_gt_150_M0=data0.loc[data0['condition'] == 'twbga_gt_150_M0']

dict={}
for p in phenotypes:
    dict[p]=data0.loc[data0['condition'] == p]

for i in phenotypes:
    for j in phenotypes:
        if i>j:
            print i,j,'phenotype.'+i+"_"+j+'.'+day+'.DE.csv'
            index=i+"_"+j
            file=open("Run"+index+day+".sh","w")
            file_phenotype="phenotype."+index+"."+day+".DE.csv"
            !ls -l $file_phenotype
            print file_phenotype
            dict_temp=dict[i].append(dict[j], ignore_index=True)
            dict_temp.to_csv('phenotype.'+index+'.'+day+'.DE.csv', sep='\t')
            file.write("~/project/anaconda2/bin/Rscript de.R "+file_phenotype+" $PWD/analysis.DE.genes."+day+"."+index+day)
            file.close()
dict_temp.head()
```
